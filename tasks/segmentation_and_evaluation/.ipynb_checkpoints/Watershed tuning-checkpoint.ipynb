```python
from __future__ import print_function

import glob
import itertools
import os
import time

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline

sns.set_style("whitegrid")
```

```python
csv_paths = glob.iglob("/nrs/turaga/grisaitisw/affinities/*/*/evaluation*csv")
dfs = [pd.read_csv(p) for p in csv_paths]
df = pd.concat(dfs)

df.rename(
    columns={
        "V_Rand_merge": "rand_merge_volumetric",
        "V_Rand_split": "rand_split_volumetric",
        "V_Info_merge": "vi_merge_volumetric",
        "V_Info_split": "vi_split_volumetric",
    },
    inplace=True
)

df.reset_index(drop=True, inplace=True)

def rand_total(rand_merge, rand_split):
    return 2.0 * rand_merge * rand_split / (rand_merge + rand_split)

df["rand_volumetric"] = rand_total(df["rand_merge_volumetric"], df["rand_split_volumetric"])
df["vi_volumetric"] = df["vi_merge_volumetric"] + df["vi_split_volumetric"]
```

```python
model_stacks = {
    "2017.04.17.2": "CUDA 8.0, cudnn6 + libdnn, caffe 2017-04-17",
    "2017.04.17.1": "CUDA 8.0, cudnn6, caffe 2017-04-17",
    "run_1208_6": "CUDA 8.0, cudnn5 + libdnn, caffe 2016-12-05",
    "2017.03.30.1": "CUDA 8.0, cudnn5 + libdnn, caffe 2016-12-05",
}

def get_stack(row):
    return model_stacks.get(row["model"], "")

df["stack"] = df.apply(get_stack, axis=1)
```
